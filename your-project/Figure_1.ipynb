```python
import os
import pandas as pd
import numpy as np
import datetime
from datetime import time
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
figure_1 = pd.read_csv("../Data/renfe_figure_1.csv")
```

```python
figure_1.head()
```

```python
from matplotlib.pyplot import style
plt.style.use('seaborn-dark-palette')
```

```python
plt.figure(figsize=(14,6))
plt.bar(figure_1["route"], figure_1["avg_daily_trains"], color="tan", edgecolor="steelblue", linewidth="3")

plt.xlabel("routes", fontsize = 14)
plt.ylabel("# departing trains", fontsize = 14)
plt.tick_params(axis="x", labelsize=12)
plt.tick_params(axis="y", labelsize=12)


plt.title("Daily average departing trains", fontsize = 15)

plt.savefig("../Figures/Daily average departing trains", dpi=300)
plt.show()
```
