```python
import os
import pandas as pd
import numpy as np
import datetime
from datetime import time
```

#### Just keeping the number of the calculated column days to departure

```python
renfe["days_to_departure"] = renfe["days_to_departure"].apply(lambda x: x[:2])
```

### Reducing train type categories into 3 big categories

```python
renfe.loc[renfe["train_type"].str.match("AVE"), "train_category"] = "High_speed"
renfe.loc[renfe["train_type"].str.match("AVE-TGV"), "train_category"] = "High_speed"
```

```python
renfe.loc[renfe["train_type"].str.match("MD"), "train_category"] = "Regional"
renfe.loc[renfe["train_type"].str.match("LD"), "train_category"] = "Regional"
renfe.loc[renfe["train_type"].str.match("EXPRESS"), "train_category"] = "Regional"
renfe.loc[renfe["train_type"].str.match("REGIONAL"), "train_category"] = "Regional"
```

```python
renfe.loc[renfe["train_type"].str.match("INTERCITY"), "train_category"] = "Long_distance"
renfe.loc[renfe["train_type"].str.match("ALVIA"), "train_category"] = "Long_distance"
renfe.loc[renfe["train_type"].str.match("TRENHOTEL"), "train_category"] = "Long_distance"
renfe.loc[renfe["train_type"].str.match("AV City"), "train_category"] = "Long_distance"
```

### Reducing ticket classes into 2 big categories

```python
renfe.loc[renfe["ticket_class"].str.match("Turista"), "ticket_class_category"] = "Economy"
renfe.loc[renfe["ticket_class"].str.match("Preferente"), "ticket_class_category"] = "First_class"
renfe.loc[renfe["ticket_class"].str.match("Clase"), "ticket_class_category"] = "First_class"
```

### Reducing fare types into 2 big categories

```python
renfe.loc[renfe["fare"].str.match("Promo"), "fare_category"] = "Fixed"
renfe.loc[renfe["fare"].str.match("Promo +"), "fare_category"] = "Fixed"
renfe.loc[renfe["fare"].str.match("Adulto"), "fare_category"] = "Fixed"
renfe.loc[renfe["fare"].str.match("Mesa"), "fare_category"] = "Fixed"
renfe.loc[renfe["fare"].str.match("Grupos Ida"), "fare_category"] = "Fixed"

renfe.loc[renfe["fare"].str.match("Flexible"), "fare_category"] = "Flexible"
```
